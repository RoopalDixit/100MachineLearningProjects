```python
from ultralytics import YOLO
import torch

# Load pre-trained YOLOv8 nano model
model = YOLO("yolov8n.pt")

# Train on your dataset
results = model.train(
    data="/Users/roopaldixit/Desktop/100Projects/ObjectDetection_YOLO/final_data/data_wl.yaml",
    epochs=1,                    # 50 epochs for initial training
    imgsz=640,                    # Standard input size
    batch=16,                     # Adjust if memory issues occur
    name="wildlife_yolo",         # Output folder: runs/detect/train/wildlife_yolo
    device=0 if torch.cuda.is_available() else "cpu"  # GPU or CPU
)

# Print training results
print(f"Training complete! Metrics: {results}")
```

```python

metrics = model.val(data="/Users/roopaldixit/Desktop/100Projects/ObjectDetection_YOLO/final_data/data_wl.yaml")
print(f"mAP@50: {metrics.box.map50:.3f}")
print(f"mAP@50:95: {metrics.box.map:.3f}")
```

```python
from ultralytics import YOLO
import cv2
import os

# Load trained model
model = YOLO("/Users/roopaldixit/Desktop/100Projects/ObjectDetection_YOLO/yolov8n.pt")

# Test images directory
test_dir = "/Users/roopaldixit/Desktop/100Projects/ObjectDetection_YOLO/final_data/test/images"
output_dir = "wildlife_outputs"
os.makedirs(output_dir, exist_ok=True)

# Run inference
results = model(test_dir, save=True, conf=0.5)  # conf=0.5 filters low-confidence detections

# Process and save annotated images
for result in results:
    annotated_img = result.plot()  # Draws boxes/labels
    # Save (Ultralytics auto-saves to runs/detect/predict/, but we can customize)
    img_name = result.path.split('/')[-1]
    cv2.imwrite(os.path.join(output_dir, f"detected_{img_name}"), annotated_img)
    print(f"Processed {img_name}: {len(result.boxes)} objects detected")

    # Print details
    if result.boxes is not None:
        for box in result.boxes:
            cls = int(box.cls[0])
            conf = box.conf[0]
            class_name = model.names[cls]
            print(f"  - {class_name}: {conf:.2f}")
```
